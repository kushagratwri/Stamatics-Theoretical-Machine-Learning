```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0, 10, 50)
epsilon = np.random.normal(0, np.sqrt(5), 50)
```

```python
y = 3 * x + 5 + epsilon
plt.scatter(x, y)
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
xy_mean= np.mean(x*y)
x2_mean = np.mean(x**2)
y2_mean = np.mean(y**2)
```

```python
b1= (xy_mean - x_mean * y_mean) / (x2_mean - x_mean**2)
b0 = (y_mean *x2_mean - x_mean * xy_mean) / (x2_mean - x_mean**2)
```

```python
y_predicc= b0 + b1 * x
plt.scatter(x,y, color= 'blue')
plt.plot(x, y_predicc, color='red')
plt.show()
```
